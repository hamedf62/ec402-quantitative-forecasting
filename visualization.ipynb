```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--b',ms = 20,mec = 'r',linewidth = '5')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
y3 = np.array([7, 3, 6, 18])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.xlabel("this is X axis")
plt.ylabel("this is Y axis")

font1 = {"family": "tahoma", "color": "purple", "size": 50}

plt.title("this is my first graph", fontdict=font1)

plt.show()
```
